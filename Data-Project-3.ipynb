```python
#Create a linegraph of Google adjucted close values
fig, ax1 = plt.subplots()

x_axis = google['dates']
y_axis = google["adjusted close"]

markers_on = ["2023-02-28","2023-03-31", "2023-04-28", "2023-05-31", "2023-09-29", "2023-10-31", "2023-12-29" ]
GoogleChange, = plt.plot(x_axis, y_axis, linewidth=1, color='r', label="Google Stock")
plt.title("Google Value Over Time")
plt.xlabel("Date")
plt.ylabel("Closing Value (adjusted)")

for marker_date in markers_on:
        index = x_axis[x_axis.dt.strftime('%Y-%m-%d') == marker_date].index[0]
        value = y_axis.loc[index]
        ax1.annotate(f'Marker', xy=(index, value), xytext=(index, value + 10),
                 arrowprops=dict(facecolor='red', arrowstyle='->'),
                 fontsize=10, color='red', ha='center', va='bottom')

plt.legend(handles=[GoogleChange], loc="upper right")

fig.tight_layout()
plt.show()
```
